```python
import pandas as pd

# Load
df = pd.read_csv("/content/combined_fictional_financial_transactions.csv")
print("Loaded:", df.shape)
```

```python
# Basic cleaning
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
df['Amount'] = df['Amount'].astype(str).str.extract(r'(\d+\.?\d*)').astype(float)
df['Fee'] = df['Fee'].astype(str).str.extract(r'(\d+\.?\d*)').astype(float)
```

```python
# Drop missing or duplicate rows
df = df.dropna(subset=['Date','Amount']).drop_duplicates()
```

```python
# Keep only spending-related rows
df = df[df['Type'].isin(['Purchase','Payment','Transfer'])]
```

```python
# Feature engineering (simple)
df['Net_Amount'] = df['Amount'] + df['Fee']      # total outflow
df['Week'] = df['Date'].dt.isocalendar().week
df['Day']  = df['Date'].dt.day_name()
```

```python
# Save for submission
df.to_csv("financial_transactions_clean.csv", index=False)
print("✅ Saved cleaned file:", df.shape)
```
